```python
import pandas as pd

deliveries = pd.read_csv('deliveries.csv')
complaints = pd.read_csv('complaints.csv')
customers = pd.read_csv('customers.csv')
drivers = pd.read_csv('drivers.csv')
hubs = pd.read_csv('hubs.csv')
orders = pd.read_csv('orders.csv')
vehicles = pd.read_csv('vehicles.csv')
incidents = pd.read_csv('incidents.csv')
app_events = pd.read_csv('app_events.csv')
```

```python
deliveries.head()
```

## Initial Dataset Check

This section checks the size of each dataset to confirm that all CSV files were loaded correctly.

```python
datasets = {
    "deliveries": deliveries,
    "complaints": complaints,
    "customers": customers,
    "drivers": drivers,
    "hubs": hubs,
    "orders": orders,
    "vehicles": vehicles,
    "incidents": incidents,
    "app_events": app_events
}

for name, df in datasets.items():
    print(f"{name}: {df.shape[0]} rows, {df.shape[1]} columns")
```

## Column Structure and Missing Values

This section examines the column names and checks for missing values in each dataset.

```python
for name, df in datasets.items():
    print(f"\n{name.upper()} COLUMNS:")
    print(df.columns.tolist())

    print("\nMissing values:")
    print(df.isnull().sum())
```
